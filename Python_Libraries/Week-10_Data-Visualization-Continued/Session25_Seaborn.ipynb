### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level function -> rectangular shape of figure
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size') # hue segregate datapoints by color, size by size and style of shapes
```

```python
# scatter using relplot -> size and hue
# relplot -> figure level function-> square shape of graph
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size') # style semantics
```

## Scatter plot library recommends that we have to use figure level function.
## relplot is general it can plot both scatter as well as line plot.

```python
# line plot : In Time series data we prefer line plot
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function -> rectangle shape graph
# life expentencies of India over the year
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relpplot -> figure level function -> square shape graph
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

### We can observe below points from above 2 graphs:
- In figure level function, legend is shown outside the actual graph and in axis level function, legend is shown inside the actual graph.

- This is because figure level function have the control over the whole figure and axis level function have the control over a axis only.

```python
# facet plot -> figure level function -> work with relplot
# Using facet we can plot multiple scatter plot based on a single column
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='day')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year')
```

```python
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```


## It will only work with figure level function (relplot), will not work with axis level graph (scatterplot and lineplot)
## Because figure level function have the control over whole figure.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', col='sex', row='day')
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist',bins=2)
```

```python
# Its also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# histogram is acting like a countplot here
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# element -> step (only outline, no bars will be there)
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist',element='step',hue='sex')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python
# kdeplot (kernel density estimation)
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

### Kde plot tells the probability of values.
If we have some missing value in between then it will be visualize by the kde plot easily.

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python
# Rugplot : It is a different kind of plot visible on x-axis
# We can determine the density of the data plots using this visualization

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# we can observe the combination of column values
# sns.histplot(data=tips, x='total_bill', y='tip') # same result
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip') # closer live have high density and spacious lines have less density
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
gap # long data
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp') # wide data
temp_df
```

```python
# axes level function
plt.figure(figsize=(15,15)) # we can resize axes level function in similar way as matplotlib
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer') # linewidth: gap b/w boxes, cmap: color, annot: fill boxes with value
```

```python
# Clustermap : This brings similar columns together and dissimilar columns together

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]]) # sending all 4 columns
```

# Observations
- It organise  columns as per its preference.
- sepal_width and petal_length is similar and sepal_length and petal_width is dissimilar as per our data
- Dendogram (hierarchical clustering) is visible in side of the above graph
